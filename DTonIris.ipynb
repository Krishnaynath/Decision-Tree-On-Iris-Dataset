```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
import pandas as pd
```

- Imports the pyplot module from matplotlib and gives it the alias plt.
- Loads the famous Iris dataset from scikit-learn’s built-in datasets.
- Imports the function train_test_split that splits data into training and testing sets.
- Imports for decision tree and plot tree.
- Imports the pandas library, a powerful tool for handling datasets.

```python
# Load dataset
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = iris.target
```

- Loads the Iris dataset into a variable called iris.
- Creates a DataFrame (a table) using pandas.
- iris.target gives the class labels as numbers.

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

Function from scikit-learn that splits your dataset into 2 parts:

  - Training Set
  - Testing Set

```python
# Build Decision Tree
clf = DecisionTreeClassifier(criterion="entropy", random_state=42)
clf.fit(X_train, y_train)
```

- DecisionTreeClassifier → This creates a Decision Tree model for classification tasks.
- Parameters used:
   - criterion="entropy"
   - random_state=42
- clf.fit(X_train, y_train) -> trains the model.

```python
# Plot decision tree
plt.figure(figsize=(12,8))
plot_tree(clf, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)
plt.title("Decision Tree on Iris Dataset")
plt.show()
```

- Creates a new figure for plotting.
- plot_tree → function from sklearn.tree to draw the decision tree.
- plt.title("Decision Tree on Iris Dataset") -> Adds a title above the plot.
- plt.show() -> Actually renders the plot so you can see the decision tree diagram.

```python
# Plot feature importance (Information Gain)
importances = clf.feature_importances_
features = iris.feature_names
```

- clf.feature_importances_ = numerical importance values (information gain contribution).

- iris.feature_names = labels for those features.

Together, they create a graph showing which features are most important for classification.

```python
plt.figure(figsize=(8,5))
plt.bar(features, importances, color='skyblue')
plt.xlabel("Features")
plt.ylabel("Information Gain (Feature Importance)")
plt.title("Information Gain of Features in Iris Dataset")
plt.show()
```

- Creates a new figure for the bar chart.
- Draws the bar chart.
- Labels the x-axis as Features.
- Labels the y-axis as Information Gain (Feature Importance).
- Adds a title above the chart.
- Displays the bar chart.
